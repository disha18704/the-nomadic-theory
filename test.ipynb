```python
import re 

pattern = re.compile(r'^(?:[456]\d{15}|[456]\d{3}-\d{4}-\d{4}-\d{4})$')


def is_valid_cc_num(s):
    
    if not pattern.match(s):
        return False
    digits = s.replace('-','')
    
    if re.search(r'(\d)\1{3}',digits):
        return False
    return True
    
N = int(input().strip())
for _ in range(N):
    s = '5122-2368-7954-3214'
    if is_valid_cc_num(s):
        print("Valid")
    else:
        print("Invalid")
    
```

```python
def is_valid_code(s):
    if len(s) != 10:
        return False
    if not s.isalnum():
        return False
    
    upper_count = sum( 1 for ch in s if ch.isupper())
    if upper_count < 2:
        return False
    
    digit_count = sum(1 for ch in s if ch.isdigit())
    if digit_count < 3:
        return False
    
    if len(set(s)) != len(s):
        return False
    
    return True
    





if __name__ == '__main__':
    N = int(input().strip())
    for _ in range(N):
        s = input().strip()
        if is_valid_code(s):
            print("Valid")
        else:
            print("Invalid")
```

```python
import re

def is_valid_access_code(s):
    if len(s) != 10 or not s.isalnum():
        return False
    
    if len(s) != len(set(s)):
        return False
    
    upper_char_count = len(re.findall(r'[A-Z]',s))
    if upper_char_count < 2:
        return False
    
    digit_count = len(re.findall(r'\d',s))
    if digit_count < 3:
        return False
    
    return True


if __name__ == '__main__':
    N = int(input().strip())
    for _ in range(N):
        s = input().strip()
        if is_valid_access_code(s):
            print("Valid")
        else:
            print("Invalid")
```
